```python
#reading csv file and converting into pandas dataframe
#C:\Users\dtawade\zfscsv

acs_df = pd.read_excel('C:\\Users\\dtawade\\ybl-data.xlsx')

# calling head of data frame to see few data

acs_df.head()
```

```python
# using catplot to plot hours worked i.e. quantity, 
# This gives you and idea about the hours entered in your project and the no of entries.
# If we see below dist, we see that max people have entered 8 hours.
# The 3 hour distribution is for the 2nd shift which is split into 5 hours normal time and 3 hours OT
# from 1pm to 6pm 5hrs and from 6pm to 9pm 3hr evening OT.
# we have close to 25 entries with 6 hours

sns.catplot('Quantity',kind='count',data=acs_df,height=10)
```

```python
#creating new data from original excel data frame acs_df with entries for time entered as 6 hours
# as 6 hours is not a general shift split
newdf1 = acs_df[(acs_df.Quantity == 6)]

newdf1.head(3)
```

```python
# in the 6 hour data frame querying the task numbers entered
sns.catplot('Task Number',kind='count',data=newdf1,hue="Person Email",height=10)
```
